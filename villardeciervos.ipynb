```python
# Instalar paquetes necesarios
!pip install fpdf2
!pip install markdown2

# Importar librerías
from fpdf import FPDF
from datetime import datetime
from google.colab import files

# Configurar PDF con fuente Unicode
pdf = FPDF()
pdf.add_page()
# Use Liberation Sans font, which is usually available in Colab and supports Unicode
pdf.add_font("LiberationSans", "", "/usr/share/fonts/truetype/liberation/LiberationSans-Regular.ttf", uni=True)
pdf.add_font("LiberationSans", "B", "/usr/share/fonts/truetype/liberation/LiberationSans-Bold.ttf", uni=True)
pdf.set_font("LiberationSans", size=12)

# 1. Título del documento
pdf.set_font("LiberationSans", 'B', 16)
pdf.cell(0, 10, 'INFORME TÉCNICO Y PRESUPUESTO PARA REFORMA DE INMUEBLES', 0, 1, 'C')
pdf.set_font("LiberationSans", '', 12)
pdf.cell(0, 10, f'Villardeciervos (Zamora) - {datetime.now().strftime("%d/%m/%Y")}', 0, 1, 'C')
pdf.ln(15)

# 2. Datos de los inmuebles
pdf.set_font("LiberationSans", 'B', 14)
pdf.cell(0, 10, '1. DATOS DE LOS INMUEBLES', 0, 1)
pdf.set_font("LiberationSans", '', 10)

# Tabla de datos
col_width = pdf.w / 4.5
row_height = pdf.font_size * 2

def create_table(headers, data):
    pdf.set_font("LiberationSans", 'B', 10)
    for header in headers:
        pdf.cell(col_width, row_height, header, border=1)
    pdf.ln(row_height)

    pdf.set_font("LiberationSans", '', 10)
    for row in data:
        for item in row:
            pdf.cell(col_width, row_height, item, border=1)
        pdf.ln(row_height)
    pdf.ln(10)

# Datos principales
inmuebles_data = [
    ["Concepto", "Inmueble 1", "Inmueble 2"],
    ["Referencia Catastral", "4770905QG2447S0001QO", "4770904QG2447S0001GO"],
    ["Superficie", "117 m²", "135 m²"],
    ["Año Construcción", "1900", "1900"],
    ["Uso Principal", "Vivienda + Almacén", "Vivienda + Almacén"]
]
create_table(inmuebles_data[0], inmuebles_data[1:])

# 3. Presupuesto estimado
pdf.set_font("LiberationSans", 'B', 14)
pdf.cell(0, 10, '2. PRESUPUESTO ESTIMADO', 0, 1)

# A. Demolición
pdf.set_font("LiberationSans", 'B', 12)
pdf.cell(0, 10, 'A. Demolición', 0, 1)

demolicion_data = [
    ["Concepto", "Inmueble 1", "Inmueble 2", "Total"],
    ["Demolición total", "8.000 EUR", "9.500 EUR", "17.500 EUR"]
]
create_table(demolicion_data[0], demolicion_data[1:])

# B. Rehabilitación
pdf.set_font("LiberationSans", 'B', 12)
pdf.cell(0, 10, 'B. Rehabilitación/Reconstrucción', 0, 1)

rehab_data = [
    ["Concepto", "Inmueble 1", "Inmueble 2", "Total"],
    ["Estructura", "45.000 EUR", "50.000 EUR", "95.000 EUR"],
    ["Instalaciones", "12.000 EUR", "14.000 EUR", "26.000 EUR"],
    ["Acabados", "18.000 EUR", "20.000 EUR", "38.000 EUR"],
    ["Accesibilidad", "6.500 EUR", "7.000 EUR", "13.500 EUR"],
    ["TOTAL", "81.500 EUR", "91.000 EUR", "172.500 EUR"]
]
create_table(rehab_data[0], rehab_data[1:])

# 4. Subvenciones aplicables
pdf.set_font("LiberationSans", 'B', 14)
pdf.cell(0, 10, '3. SUBVENCIONES APLICABLES', 0, 1)

subv_data = [
    ["Concepto", "Inmueble 1", "Inmueble 2", "Total"],
    ["PRR Next Generation", "18.000 EUR", "18.000 EUR", "36.000 EUR"],
    ["Programa 11", "12.500 EUR", "12.500 EUR", "25.000 EUR"],
    ["Diputación Zamora", "9.000 EUR", "9.000 EUR", "18.000 EUR"],
    ["TOTAL", "39.500 EUR", "39.500 EUR", "79.000 EUR"]
]
create_table(subv_data[0], subv_data[1:])

# Guardar y descargar
pdf_output = "Informe_Reforma_Villardeciervos.pdf"
pdf.output(pdf_output)
files.download(pdf_output)

print("✅ PDF generado y descargado correctamente")
```
